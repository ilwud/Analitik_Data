```python
import pandas as pd

# Load the dataset
file_path = "D:/Kuliah/semester2/Analitik Data/H5 Analitik/data_quiz.xlsx"
xls = pd.ExcelFile(file_path)

# Check sheet names
xls.sheet_names
```

```python
# Read the data from the "Gabungan" sheet
df = pd.read_excel(xls, sheet_name="Gabungan")

# Display basic information about the dataset
df.info(), df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Hitung jumlah missing values
missing_values = df.isnull().sum()

# Drop baris dengan missing values
df_cleaned = df.dropna()

# Boxplot nilai Quiz berdasarkan Kelas
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned["Kelas"], y=df_cleaned["Quiz"], palette="coolwarm")
plt.xlabel("Kelas")
plt.ylabel("Nilai Quiz")
plt.title("Boxplot Nilai Quiz Berdasarkan Kelas")
plt.show()

# Menampilkan jumlah missing values sebelum pembersihan
missing_values
```

```python
# Boxplot nilai Quiz berdasarkan Gender
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned["Gender"], y=df_cleaned["Quiz"], palette="coolwarm")
plt.xlabel("Gender")
plt.ylabel("Nilai Quiz")
plt.title("Boxplot Nilai Quiz Berdasarkan Gender")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned["Kelas"], y=df_cleaned["IPS2"], palette="coolwarm")
plt.xlabel("Kelas")
plt.ylabel("Nilai IPS Semester 2")
plt.title("Boxplot Nilai IPS Semester 2 Berdasarkan Kelas")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_cleaned["Gender"], y=df_cleaned["IPS2"], palette="coolwarm")
plt.xlabel("Gender")
plt.ylabel("Nilai IPS Semester 2")
plt.title("Boxplot Nilai IPS Semester 2 Berdasarkan Gender")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_cleaned["IPS2"], y=df_cleaned["Quiz"], alpha=0.7, edgecolors="k")
plt.xlabel("Nilai IPS Semester 2")
plt.ylabel("Nilai Quiz")
plt.title("Scatter Plot Nilai Quiz dengan Nilai IPS Semester 2")
plt.show()
```

```python
correlation = df_cleaned["Quiz"].corr(df_cleaned["IPS2"], method="pearson")
print(f"Korelasi Pearson antara nilai Quiz dan IPS2: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df_cleaned["Quiz"], bins=10, kde=True, color="blue")
plt.xlabel("Nilai Quiz")
plt.ylabel("Frekuensi")
plt.title("Histogram Nilai Quiz")
plt.show()
```
